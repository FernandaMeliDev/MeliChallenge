```python
import requests
import json

# Lista de alertas/incidentes
alert_ids = range(1, 101)  # gera aleatoriamente IDs de 1 a 100

# URL da API simulada
api_url = "https://api.exemplo.com/alerts/"

# Função para buscar detalhes de cada alerta
def get_alert_details(alert_id):
    try:
        response = requests.get(f"{api_url}{alert_id}")
        if response.status_code == 200:
            return response.json()
        else:
            return {
                "error": "resource not found",
                "message": "Si quieres conocer los recursos de la API que se encuentran disponibles visita el sitio de desarrolladores de MercadoLibre (http://developers.mercadolibre.com)"
            }
    except Exception as e:
        return {"error": "request failed", "message": str(e)}

# Simula varredura
alerts_data = []
for alert_id in alert_ids:
    alert_details = get_alert_details(alert_id)
    alerts_data.append(alert_details)

# Exibindo os dados obtidos
print(json.dumps(alerts_data, indent=4))
```

```python
import csv

#Campos para analise
fields = ['alert_id', 'type_of_alert', 'status', 'assigned_to', 'creation_date', 'resolution_date', 'impact_level']

#Simula dados coletados
alerts_data = [
    {
        "alert_id": 1,
        "type_of_alert": "Compliance",
        "status": "Open",
        "assigned_to": "Team A",
        "creation_date": "2023-07-01",
        "resolution_date": None,
        "impact_level": "High"
    },
]

#Filtra dados coletados
cleaned_data = [ 
    { key: alert.get(key, '') for key in fields } 
    for alert in alerts_data if "error" not in alert
]


#Escreve CSV
with open('alerts_data.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fields)
    writer.writeheader()
    writer.writerows(alerts_data)

print("Dados salvos em 'alerts_data.csv'")
```
